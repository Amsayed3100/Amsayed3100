```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#prediction function
#input features
X=np.array([[1,9,42],
               [2,8,48],
               [3,8,52],
               [4,8,58],
               [5,7,65],
               [6,6,70]
               ])
print(X)
#target values
Y=np.array([55,60,65,70,80,87])
print(Y)
```

```python
#MakePredictions
def make_prediction(X,w,b):
  m=X.shape[0]
  predc_list=np.zeros((m,))

#Y=w*x + b
  for i in range(m):
    predc_list[i]=np.dot(w,X[i])+b

  return predc_list






```

```python
n=X.shape[1]
m=X.shape[0]
w_init=np.ones((n,))
actual=Y
print(actual)
print(make_prediction(X,w_init,b=1))
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#Calculate Cost functions
def compute_cost(X,Y,w,b):
  m=X.shape[0] #row
  cost=0.0

  #prediction
  predict_lst=make_prediction(X,w,b)
  #error=predt_value-actual_value
  error=(predict_lst-Y)
  #error_square
  error_square=error**2
  #sumerror_square
  sum=np.sum(error_square)
  #defineMSECost
  MSE_Cost=sum/m

  return MSE_Cost



```

```python
w=np.ones((n,))
compute_cost(X,Y,w,b=1.00)
```

```python
def calculate_gradient(X,Y,w,b):
  m=X.shape[0] #row
  n=X.shape[1] #column
  dj_dw=np.zeros((n,))
  dj_db=0.0
  for i in range(m):
    predict=np.dot(w,X[i])+b
    error=predict-Y[i]
    dj_db=dj_db+error
    for j  in range(n):
      dj_dw[j]=dj_dw[j]+(error*X[i,j])



  return dj_dw/m,dj_db/m



```

```python
n=X.shape[1]
w=np.ones((n,))

calculate_gradient(X,Y,w,b=1.0)
```

```python
#Gradient decent
def gradient_decent(X,Y,w_init,b_init,iterator,alpha):
  w=w_init
  b=b_init


  for i in range(iterator):
    dj_dw,dj_db=calculate_gradient(X,Y,w,b)

    w=w-(alpha*dj_dw)
    b=b-(alpha*dj_db)
    cost=compute_cost(X,Y,w,b)

    if i%100==0:
      print(f"itreration:{i},Cost:{cost:0.4f},W:{w},b:{b}")

  return w,b


```

```python
n=X.shape[1]
w=np.zeros((n,))
alpha=0.01
iterator=10000
b=1.00

gradient_decent(X,Y,w,b,iterator,alpha)
```
